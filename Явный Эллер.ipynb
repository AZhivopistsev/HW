```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
N = 300
k = 0.01
w = 1.5
t = np.linspace(-10, 10, N)
dt = t[1] - t[0]
x = np.zeros(N)
x[0] = 0.5
x1 = np.zeros(N)
for i in range(N-1):
    x1[i + 1] = x1[i] + dt * (k * x[i]**3 - w**2 * x[i])
    x[i + 1] = x[i] + dt * x1[i]
 
plt.plot(t, x)
plt.grid()
plt.show()
```
