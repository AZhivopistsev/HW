```python
import matplotlib.pyplot as plt
import numpy as np

N = 10000
k = 0.1
w = 6
t = np.zeros(N)
dt = 0.0001
x = np.zeros(N)
x[0] = 2
x1 = np.zeros(N)
for i in range(N - 1):
    t[i + 1] = t[i] + dt
    x1[i + 1] = x1[i] + dt * (k * x[i]**3 - w**2 * x[i])
    x[i + 1] = x[i] + dt * x1[i]
 
plt.plot(t, x)
plt.grid()
plt.show()
```
