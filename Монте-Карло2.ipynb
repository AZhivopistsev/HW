```python
from numpy import random
from numpy import linspace
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import mlab
%matplotlib notebook

N = 1000000
M1 = 40
sM1 = 2
M2 = 30
sM2 = 10
R = 3.2
sR = 1
m1 = random.normal(M1, sM1, size=N)
m2 = random.normal(M2, sM2, size=N)
r = random.normal(R, sR, size=N)
res = 0
for i in range(len(r)):
    if(r[i] <= 0):
        res += 1
print("Значений радиуса меньше 0 или равных:", res)
F = 6.67384 * m1[np.where(r!=0)] * m2[np.where(r!=0)] / (r[np.where(r!=0)] * r[np.where(r!=0)]) * 1000

f = 1000 * M1*M2*6.67384/(R*R)
sf = f * ((sM1/M1)**2 + (sM2/M2)**2 + 4 * (sR/R)**2)**0.5
plt.figure()
plt.hist(F,bins = 10, normed = True)
#x = linspace(f - 3 * sf, f + 3 * sf, 1000)
#plt.plot(x, mlab.normpdf(x, f, sf))
plt.show()
```
