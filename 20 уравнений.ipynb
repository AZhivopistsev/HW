```python
def function():
    import numpy as np
    import random
    n = 20
    mas = np.zeros((n,n+1))
    for i in range(n):
        if(i != 0):
            mas[i,i-1] = random.randint(0,20)
        mas[i,i] = random.randint(0,20)
        if(i != n-1):
            mas[i,i+1] = random.randint(0,20)
        mas[i,n] = random.randint(0,20)
    print(mas)    
    for i in range(n - 1):
        mas[i+1, 0:] = mas[i+1, 0:] - mas[i,0:]/mas[i,i]*mas[i+1,i]
    
    for j in range(n-1, 0, -1):
        mas[j-1,0:] = mas[j-1,0:] - mas[j,0:]*mas[j-1,j]/mas[j,j]
    for x in range(n):
        print(mas[x,n]/mas[x,x])
    
    
function()
```
